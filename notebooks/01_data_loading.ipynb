## Step 4B: Fix Missing Values — Matches Dataset

Based on our missing value analysis, decisions made:

| Column | Missing | % | Decision | Fill Value |
|--------|---------|---|----------|------------|
| city | 51 | 4.7% | Fill | 'Unknown' |
| player_of_match | 5 | 0.5% | Fill | 'Unknown' |
| winner | 5 | 0.5% | Fill | 'No Result' |
| result_margin | 19 | 1.7% | Fill | 0 |
| target_runs | 3 | 0.3% | Fill | 0 |
| target_overs | 3 | 0.3% | Fill | 0 |
| method | 1074 | 98% | **DROP** | Too many missing! |

**Rule applied:**
- Missing < 50% → Fill with appropriate value
- Missing > 50% → Drop the column

```python
print("============"*50)

print("Before handling missing values in Matches dataset...")
print(df1.shape)
print(df1.isnull().sum())
df1['city']= df1['city'].fillna('Unknown')
df1['player_of_match']=df1['player_of_match'].fillna('Unknown')
df1['winner']=df1['winner'].fillna('No Result')
df1['result_margin']=df1['result_margin'].fillna(0)
df1['target_runs']=df1['target_runs'].fillna(0)
df1['target_overs']=df1['target_overs'].fillna(0)
df1 = df1.drop(columns=['method'])
print("After handling missing values, Matches dataset info:")
print("============"*50)
print(df1.shape)
print(df1.isnull().sum())
```

## Step 4C: Fix Missing Values — Deliveries Dataset

| Column | Missing | % | Decision | Reason |
|--------|---------|---|----------|--------|
| extras_type | 246,795 | 94.6% | **DROP** | Too many missing |
| fielder | 251,566 | 96.4% | **DROP** | Too many missing |
| player_dismissed | 247,970 | 95.1% | Fill 'not_out' | NaN = batsman survived! |
| dismissal_kind | 247,970 | 95.1% | Fill 'not_out' | NaN = not dismissed! |

**Special Rule Applied:**
> Missing > 50% normally means DROP
> BUT player_dismissed and dismissal_kind are SPECIAL!
> NaN here means batsman was NOT OUT that ball
> This is MEANINGFUL information — not truly missing!
> So we KEEP and fill with 'not_out' label

```python
print("============"*50)
print("Before handling missing values in Deliveries dataset...")
print(df2.shape)
print(df2.isnull().sum())


df2=df2.drop(columns=['fielder'])
df2=df2.drop(columns=['extras_type'])
df2['player_dismissed']=df2['player_dismissed'].fillna('not out')
df2['dismissal_kind']=df2['dismissal_kind'].fillna('not out')
print("After handling missing values, Deliveries dataset info:")
print("============"*50)
print(df2.shape)
print(df2.isnull().sum())
```

## Step 5: Save Cleaned Data

Saving both cleaned datasets to `data/processed/` folder.
